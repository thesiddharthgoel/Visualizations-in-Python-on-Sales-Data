```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sales = pd.read_csv("SalesData.csv")
```

```python
sales.head()
```

1. Compare Sales by region for 2016 with 2015 using bar chart

```python
sales_region = round(sales.groupby("Region")[["Sales2015", "Sales2016"]].sum(), 2)
sales_region
```

```python
plt.figure(figsize=(10,5))
sales_region.plot(kind="bar")
plt.ylabel("Sales")
plt.title("Sales by region for 2016 with 2015")
plt.show()
```

```python
print("From the bar chart, we can say that sales in year 2016 is more as compared to sales in year 2015 in all regions. Also we can say that East region has contributed the maximum")
```

2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as - </br>
 Jan - Mar : Q1</br>
 Apr - Jun : Q2</br>
 Jul - Sep : Q3</br>
 Oct - Dec : Q4 

```python
sales.head()
```

```python
qtr = []
for i in sales["Month"]:
  if i in ["Jan", "Feb", "Mar"]:
    qtr.append("Q1")
  elif i in ["Apr", "May", "June"]:
    qtr.append("Q2")
  elif i in ["Jul", "Aug", "Sep"]:
    qtr.append("Q3")
  else:
    qtr.append("Q4")  
```

```python
sales["Qtr"] = pd.Series(qtr)
```

```python
sales.head()
```

7. Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
qtr_sales = round(sales.groupby("Qtr")[["Sales2015", "Sales2016"]].sum(), 2)
qtr_sales
```

```python
qtr_sales.plot(kind="bar", figsize=(10,9))
plt.ylabel("Sales")
plt.xlabel("Quater wise sales in year 2015 & 2016")
plt.show()
```
